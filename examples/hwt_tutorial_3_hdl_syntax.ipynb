```python
# jupyter_code_format is just shortcut cunction for pygments syntax highligh
from jupyter_code_format import jupyter_code_format

# an example component
from hwtLib.examples.showcase0 import Showcase0
# a library which can find a file where object is declared (part of standard python installation)
import inspect

def read_file(file_name):
    with open(file_name) as f:
        return f.read()

# A hwt component Showcase0 shows most of the syntax of hwt
jupyter_code_format(read_file(inspect.getfile(Showcase0)), "python")
```

```python
from hwt.synthesizer.utils import to_rtl_str
from hwt.serializer.hwt import HwtSerializer
from hwt.serializer.vhdl import Vhdl2008Serializer
from hwt.serializer.verilog import VerilogSerializer
from hwt.serializer.systemC import SystemCSerializer
for (serializer_cls, lang_name) in [
        (Vhdl2008Serializer, 'vhdl'),
        (VerilogSerializer, 'verilog'),
        (HwtSerializer, 'python'),
        (SystemCSerializer, 'c'),]:
    print(f"Showcase0 translated using {serializer_cls.__name__}")
    jupyter_code_format(to_rtl_str(Showcase0(), serializer_cls=serializer_cls), lang_name)
```
